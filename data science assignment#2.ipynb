```python
attack_mean = data["Attack"].mean()
```

```python
def set_attack(value):
    if(value > attack_mean ):
        return "High_Attack"
    elif(value == attack_mean):
        return "Attact_Neutral"
    else:
        return "Low_Attack"
```

```python
data["Attack_high_low"] = data["Attack"].apply(set_attack)
data
```

# Find speed_low and speed_high


```python
Speed_mean = data["Speed"].mean()
```

```python
def Check_speed(value):
    if(value > Speed_mean):
        return "High_Speed"
    elif(value == Speed_mean):
        return "Neutral_Speed"
    else:
        return "Low_Speed"
```

```python
data["Speed_High_Low"] = data["Speed"].apply(Check_speed)
```

# Find low_HP and High_HP

```python
HP_mean = data["HP"].mean()
```

```python
def Check_HP(value):
    if(value > HP_mean):
        return "High_HP"
    elif(value == HP_mean):
        return "Neutral_HP"
    else:
        return "Low_HP"
```

```python
data["HP_High_Low"] = data["HP"].apply(Check_HP)
```
